```python
#Question2:
#Answer:

"""In Bokeh, glyphs are the visual elements or graphical marks that represent individual data points or groups of data points in a plot.
They can take the form of various shapes, such as circles, squares, triangles, lines, bars, patches, etc.
 and are used to visually encode the data being plotted. Glyphs are fundamental building blocks in Bokeh and are used to create various types of visualizations."""

x = [1,2,3,4]
y = [5,6,7,8]
from bokeh.plotting import figure , output_file , show
p = figure(title = "bokeh plot")
output_file = ("abc.html")
p.circle(x,y , color = "blue")
p.line(x,y , color = "green")
p.square(x,y , fill_color = "green")
p.triangle(x,y , fill_color = "red")
show(p)
```

```python
#Question3:
#Answer:

x = [1,2,3,4]
y = [5,6,7,8]
from bokeh.plotting import figure , output_file , show
p = figure(title = "bokeh plot")
output_file = ("abc.html")
p.circle(x,y , legend_label = "data points")
p.xaxis.axis_label = "x axis"
p.yaxis.axis_label = "y axis"
show(p)
```
